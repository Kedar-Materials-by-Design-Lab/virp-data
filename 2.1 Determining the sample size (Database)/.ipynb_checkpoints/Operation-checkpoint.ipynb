## Yamane sampling, HEA

```python
df_scrambam = pd.read_csv("database.csv")

# Define Nsize values
Nsize_values = np.linspace(20,1300,50)

this_run = "S.21.1"
run_path = GetPath(df_scrambam, this_run)

# Storage for results
Nsize_list = []
median_series = []
p10_series = []  # 10th percentile
p90_series = []  # 90th percentile
all_points = []  # To store all (Nsize, Density) pairs for scatter plot

# Iterate over Nsize values
for Nsize in Nsize_values:
    Nsize = int(Nsize)
    density_values = []

    # Iterate over random_state from 1 to 70
    for random_state in range(1, 70):
        series1 = ExpectationValues(run_path / "virtual_properties.csv", 300, N=Nsize, random_state=random_state)["Density"]
        density_values.append(series1)
        all_points.append((Nsize, series1))  # Store all points

    # Compute statistics
    density_values = np.array(density_values)
    median_density = np.median(density_values)
    p10_density = np.percentile(density_values, 10)
    p90_density = np.percentile(density_values, 90)

    # Store results
    Nsize_list.append(Nsize)
    median_series.append(median_density)
    p10_series.append(p10_density)
    p90_series.append(p90_density)

# Convert all points to arrays for plotting
all_Nsize, all_density = zip(*all_points)

# Print Interdecile range at Yamane 5%
print("Interdecile range at Yamane size ("+str(Nsize_list[10])+"): "+str(p90_series[10]-p10_series[10]))
```

```python
# colors
(color1,color2) = ("teal", "cyan")

# Plot density vs Nsize with all points, median, percentiles, and shaded IQR region
plt.figure(figsize=(4, 4))
plt.scatter(all_Nsize, all_density, color="dimgrey", alpha=0.5, label="Boltzmann average (300K)", s=10)  # Small scatter points
plt.plot(Nsize_list, median_series, marker='o', label="Median", color=color1, linestyle="-")

# Shaded region (IQR, like a box in a box plot)
plt.fill_between(Nsize_list, p10_series, p90_series, color=color2, alpha=0.5, label="Interdecile Range")

# Yamane points
plt.axvline(x=286, color=color1, linestyle="--", label=None)
plt.text(300, min(p90_series)-0.0009, r"e=5%", color=color1, verticalalignment='top', fontsize=11)
plt.axvline(x=526, color=color1, linestyle="--", label=r"Yamane sample size $n_{Y}$")
plt.text(540, min(p90_series)-0.0008, r"e=3%", color=color1, verticalalignment='top', fontsize=11)

plt.xlabel("Sample size")
plt.xlim(0, 1350)
plt.ylabel("Density (g/cm³)")
plt.legend()
plt.title("HEA, PdRhNbZrSc")
plt.grid(True)

ax = plt.gca()  # Get current axis
for spine in ax.spines.values():
    spine.set_linewidth(1.3)  # Adjust the thickness as needed

plt.savefig("Yamane_HEA.png", bbox_inches='tight', dpi=300)
plt.show()
```
